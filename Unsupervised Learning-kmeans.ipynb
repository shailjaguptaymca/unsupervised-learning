```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import classification_report
```

```python
x,y=make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(x[:,0],x[:,1])
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_) # inertia is sse that we are trying to plot
    # it is trying to find the distance from the centroid off all the points. This is done after one loop that is all the 
    # iteration ( no. of times centroid is updating) has been run. i.e after one complete loop where the centroid is
    # n_int is no. of initial centroid point it will put at start)
plt.plot(range(1,11),wcss)
plt.show()
wcss
    
```

```python
#you can take 4 or 3 as the value of k for k means. As 4 would be too low  we can take k=3
kmeas = KMeans(4)
```

```python
kmeas.fit(x)
```

```python
kmeas.cluster_centers_ # this will tel all the centroids that are created
```

```python
kmeas.labels_ # all the labels for the dataset
```

```python
pred_y=kmeas.fit_predict(x)
plt.scatter(x[:,0],x[:,1])
plt.scatter(kmeas.cluster_centers_[:,0],kmeas.cluster_centers_[:,1], s=300, c='red')
plt.show()
#here kmeas is object name not kmeans as in algorithm
```
